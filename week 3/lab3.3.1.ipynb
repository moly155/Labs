```python
import numpy as np
import matplotlib.pyplot as plt
```

# Challenge 1

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
plt.plot(x,y, color='blue', label = 'x,y')
plt.plot(x,z, color='aqua', label = 'x,z')

plt.rcParams['figure.figsize'] = [12,8]

plt.ylim(0,5000)
plt.legend()
plt.title('Challenge 1')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)

axs[0].plot(x,y, color='blue', label = 'x,y')
axs[1].plot(x,z, color='aqua', label = 'x,z')

axs[0].set_title('x,y')
axs[1].set_title('x,z')


```

```python

x = np.linspace(0, 5, 100)
y = x**2
z = np.exp(x)

plt.plot(x,y, color='blue', label = 'x,y')
plt.plot(x,z, color='aqua', label = 'x,z')

plt.legend()
```

```python
plt.loglog(x,y, color='blue', label = 'x,y')
plt.loglog(x,z, color='aqua', label = 'x,z')

plt.legend()

```

#Challenge 2

```python
import pandas as pd
```

```python
fitbit = pd.read_csv('Fitbit2.csv')
fitbit
```

```python
grouped = fitbit.groupby('Months_encoded')['Steps'].mean()
df = grouped.to_frame()
df.plot(kind = 'bar', color='orange')

plt.rcParams['figure.figsize'] = [12,4]

plt.title('Mean Steps Per Month', fontsize= 15)


grouped = fitbit.groupby('Work_or_Weekend')['Steps'].mean()
df = grouped.to_frame()
df.plot(kind = 'bar', color='aqua')
plt.title('Mean Steps Work vs Weekend', fontsize= 15)



```

```python
x = fitbit['Steps']
y1 = fitbit['Minutes Very Active']
y2 = fitbit['Minutes Lightly Active']
y3 = fitbit['Minutes Sedentary']

plt.rcParams['figure.figsize'] = [8,8]

plt.scatter(x,y1, color='blue', label = 'very active')
plt.scatter(x,y2, color='aqua', label = 'fairly active')
plt.scatter(x,y3, color='purple', label = 'lightly active')

plt.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [8,8]

for name in [ 'Minutes Very Active', 'Minutes Lightly Active', 'Minutes Sedentary']:
    plt.scatter(fitbit['Steps'],fitbit[name], label = name)

plt.legend()
plt.show()
```

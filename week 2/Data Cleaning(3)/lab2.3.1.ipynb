# Challenge 2

```python
x = pd.read_csv('lab2.3.1.csv')
x
```

```python
def clean(df):
    df.columns = [col.replace(' ', '_').lower() for col in df.columns]

    def convert(cell):
        if str(cell)=='nan':
            return str(cell)
        else:
            return float(cell.replace('%',''))
        
    df['customer_lifetime_value'] = df['customer_lifetime_value'].apply(convert)

    def sex(cell):
        if 'f' in str(cell).lower():
            return 'F'
        elif 'm' in str(cell).lower():
            return 'M'
        else:
            return cell 
    

    df.gender = df.gender.apply(sex)

    df.customer_lifetime_value = df.customer_lifetime_value.astype(float)
    
    df = df.dropna(how='all')

    mean = df.customer_lifetime_value.mean()

    df['customer_lifetime_value'].fillna(mean, inplace=True)


    return df 
    

```

```python
clean(x)
```
